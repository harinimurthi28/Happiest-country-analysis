### Top 10 Happiest Countries

To find the top 10 happiest countries, we'll sort the DataFrame by 'Happiness Rank' in ascending order and then select the first 10 entries.

```python
# Sort the DataFrame by 'Happiness Rank' and display the top 10
top_10_happiest_countries = df.sort_values(by='Happiness Rank', ascending=True).head(10)
display(top_10_happiest_countries)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the aesthetic style of the plots
sns.set_style('whitegrid')

# Create a bar plot for Happiness Score of the top 10 countries
plt.figure(figsize=(12, 7))
sns.barplot(x='Country', y='Happiness Score', data=top_10_happiest_countries, palette='viridis', hue='Country', legend=False)

# Add titles and labels for clarity
plt.title('Happiness Score of Top 10 Happiest Countries', fontsize=16)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Happiness Score', fontsize=12)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add score values on top of the bars
for index, row in top_10_happiest_countries.iterrows():
    plt.text(index, row['Happiness Score'], round(row['Happiness Score'], 2), color='black', ha="center", va="bottom")

plt.tight_layout() # Adjust layout to prevent labels from overlapping
plt.show()
```
